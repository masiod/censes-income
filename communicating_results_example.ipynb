```python
import pandas as pd
%matplotlib inline 
df=pd.read_csv('./census_income_data.csv')
df.head()
```

```python
df.info()
```

```python
# group of people who got less than 50K
df_g=df[df['income'] == ' >50K']
#group of  people who got less than 50K
df_l=df[df['income'] == ' <=50K']
```

```python
df_l.head()
```

```python
# Get the value counts of education, we need the index to keep the order consistent for
#bar chart  group of people who got more than 50K
ind = df_g['education'].value_counts().index
df_g['education'].value_counts()[ind].plot(kind='bar');
```

```python
# Use the same index to keep the order consistent
# group of people who got less than 50K
df_l['education'].value_counts()[ind].plot(kind='bar');
```

```python
#Get the value counts of workclass and create a pie chart for each group.
# Use an index again to keep values in the same order
#pie chart for workclass group for people who got more than 50K
ind=df_g['workclass'].value_counts().index
df_g['workclass'].value_counts()[ind].plot(kind='pie',figsize=(8,8))
```

```python
#pie chart for workclass group for people who got less than 50K
df_l['workclass'].value_counts()[ind].plot(kind='pie',figsize=(8,8))
```

```python
#histograms of age for each group for people who got more than 50K
df_g['age'].hist()
df_g.describe()
```

```python
#histograms of age for each group for people who got less than 50K

df_l['age'].hist()
df_l.describe()
```
